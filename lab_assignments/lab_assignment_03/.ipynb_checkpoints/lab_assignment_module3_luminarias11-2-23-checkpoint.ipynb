### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import pi, exp, sin

def y(n):
    return sin(pi* n / N) * sin(20 * pi * n / N)

y_arr = []
n_arr = np.linspace(0,1000,1000)
N = len(n_arr)
for k in n_arr:
    y_arr.append(y(k))

#print(n_arr)
#print(len(n_arr))

def dft(y):
    N = len(y_arr)
    c = np.zeros(N//2+1, complex)
    for k in range(N//2+1):
        for n in range(N):
            c[k] += y_arr[n]*exp(-2j * np.pi * k * n /N)
    return c

#print(dft(y_arr))
plt.figure(figsize = (12,4))

plt.subplot(1, 2, 1)
plt.plot(np.imag(dft(y_arr)), label= "imag")
plt.plot(np.real(dft(y_arr)), label= "real")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(abs(dft(y_arr)), label = "magnitude")
plt.legend()
plt.show()
```

```python
from numpy.fft import rfft     # Fast Fourrier Transform
ft_dow = rfft(dow) 
```

```python
print(ft_dow)
```

```python
print(len(ft_dow))

plt.figure(figsize = (12,4))

plt.subplot(1, 2, 1)
plt.plot(np.real(ft_dow), label = "real")
plt.plot(np.imag(ft_dow), label = "imaginary")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(abs(ft_dow), label = "magnitude")
plt.legend()

plt.show()
```

```python
ft_dowi10 = ft_dow.copy()
ft_dowl90 = ft_dow.copy()
ft_dowi10[:len(ft_dow)//10] = 0 # This turns the lower 90% to zero
ft_dowl90[len(ft_dow)//10:] = 0 
print(ft_dowi10)
print(ft_dowl90)
```

```python
plt.plot(abs(ft_dow))
plt.plot(abs(ft_dowi10))
plt.plot(abs(ft_dowl90))
plt.show()
```

```python
from numpy.fft import irfft
inv_ft_dow = irfft(ft_dow)
inv_ft_dowi10 = irfft(ft_dowi10)
inv_ft_dowl90 = irfft(ft_dowl90)
print(inv_ft_dow)
print(inv_ft_dowi10)
print(inv_ft_dowl90)
```

```python
plt.plot(inv_ft_dow,"-r", label = "original")
#plt.plot(inv_ft_dowi10,"-g", label = "zeroes initial 10%")
plt.plot(inv_ft_dowl90,"-b", label = "zeroes final 90%")
plt.legend(loc = "upper right")

plt.show()
```
